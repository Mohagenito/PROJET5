### REQUETE 

Ce code récupère les 50 questions les plus votées contenant le tag "python" et ayant un score supérieur à 50,
publiées entre le 1er janvier 2023 et le 31 juillet 2023, puis affiche les données principales(**date, titre, tags, score**) dans un DataFrame.

```python
# Importer les bibliothèques nécessaires
from stackapi import StackAPI
import pandas as pd
from datetime import datetime
```

```python
# Initialiser l'objet StackAPI pour Stack Overflow
SITE = StackAPI('stackoverflow')

# Définir les dates de début et de fin pour la période (par exemple, du 1er juillet 2023 au 31 juillet 2023)
date_debut = int(datetime(2023, 1, 1).timestamp())
date_fin = int(datetime(2023, 7, 31).timestamp())

# Définir les paramètres pour la requête
params = {
    'page': 1,
    'pagesize': 50,
    'order': 'desc',
    'sort': 'votes',
    'tagged': 'python',
    'min': 50,
    'fromdate': date_debut,
    'todate': date_fin,
    'site': 'stackoverflow'
}

# Effectuer la requête API
questions = SITE.fetch('questions', **params)

# Extraire les données pertinentes
donnees = []
for item in questions['items']:
    donnees.append({
        'date': item['creation_date'],
        'titre': item['title'],
        'tags': item['tags'],
        'score': item['score']
    })

# Créer un DataFrame
df = pd.DataFrame(donnees)

# Convertir les timestamps Unix en dates lisibles
df['date'] = pd.to_datetime(df['date'], unit='s')

# Afficher le DataFrame
df.head()
```

```python
# Importer les bibliothèques nécessaires
from stackapi import StackAPI
import pandas as pd
from datetime import datetime

# Initialiser l'objet StackAPI pour Stack Overflow
SITE = StackAPI('stackoverflow')

# Définir les dates pour l'ensemble de données historiques (avant le 1er janvier 2023)
date_debut_historique = int(datetime(2022, 1, 1).timestamp())
date_fin_historique = int(datetime(2022, 12, 31).timestamp())

# Définir les dates pour l'ensemble de données actuelles (après le 1er janvier 2023)
date_debut_actuel = int(datetime(2023, 1, 1).timestamp())
date_fin_actuel = int(datetime(2023, 7, 31).timestamp())

# Fonction pour récupérer les questions
def recuperer_questions(date_debut, date_fin):
    params = {
        'page': 1,
        'pagesize': 50,
        'order': 'desc',
        'sort': 'votes',
        'tagged': 'python',
        'min': 50,
        'fromdate': date_debut,
        'todate': date_fin,
        'site': 'stackoverflow'
    }
    questions = SITE.fetch('questions', **params)
    donnees = []
    for item in questions['items']:
        donnees.append({
            'date': item['creation_date'],
            'titre': item['title'],
            'tags': item['tags'],
            'score': item['score']
        })
    return pd.DataFrame(donnees)

# Récupérer l'ensemble de données historiques
df_historique = recuperer_questions(date_debut_historique, date_fin_historique)
# Récupérer l'ensemble de données actuelles
df_actuel = recuperer_questions(date_debut_actuel, date_fin_actuel)

# Convertir les timestamps Unix en dates lisibles pour les deux DataFrames
df_historique['date'] = pd.to_datetime(df_historique['date'], unit='s')
df_actuel['date'] = pd.to_datetime(df_actuel['date'], unit='s')

# Enregistrer les DataFrames en fichiers CSV
df_historique.to_csv('historique.csv', index=False)
df_actuel.to_csv('actuel.csv', index=False)

# Afficher les DataFrames
print("Ensemble de données historiques enregistré sous 'historique.csv'")
print(df_historique)

print("\nEnsemble de données actuelles enregistré sous 'actuel.csv'")
print(df_actuel)
```
